## <font color="darkblue">  Inputs

```python
cDNA_file='../data/GPCR/ClassAGPCR_cDNA.txt'
Codon_ls=['GCT','GCC','GCA','GCG','TGT','TGC','GAT','GAC','GAA','GAG','TTT','TTC','GGT','GGC','GGA','GGG','CAT','CAC','ATT','ATC','ATA','AAA','AAG','TTA','TTG','CTT','CTC','CTA','CTG','ATG','AAT','AAC','CCT','CCC','CCA','CCG','CAA','CAG','AGA','AGG','CGT','CGC','CGA','CGG','TCT','TCC','TCA','TCG','AGT','AGC','ACT','ACC','ACA','ACG','GTT','GTC','GTA','GTG','TGG','TAT','TAC']
Query='ADRB2_NM_000024'
Numbering=pd.read_csv('../data/GPCR/Residue_Generic.csv',index_col='Residue').to_dict()['Generic']

HighCovPosi_DF=pd.read_csv('../data/GPCR/ClassAGPCR_HigCovPosi.csv',index_col='index')
display(HTML(HighCovPosi_DF.head(5).to_html()))
HighCovPosi_dict=HighCovPosi_DF.T.to_dict('list')
Posi_queryMatch_r=HighCovPosi_DF.to_dict('index')
Posi_queryMatch=dict()
for g in Posi_queryMatch_r: # Exchange value to key
    Posi_queryMatch[g]=dict([(str(v), str(k)) for k, v in Posi_queryMatch_r[g].items()])

TCGA_GPCR=pd.read_csv('../data/Genomics/TCGA/TCGA_panCancer_ClassAGPCR.csv')
display(HTML(TCGA_GPCR.head(5).to_html()))
Top10Mut_cancer=pd.read_csv('../data/Genomics/TCGA/TCGA_panCancer_ClassAGPCR_MutCount.csv',index_col=[0]).sort_values(by='Total.NonSyn',ascending=False).index.tolist()[:11]
print (Top10Mut_cancer)
```

## <font color="darkblue"> Compute Expected Mutation Freq by Each Codon  </font>
- Randomly select a codon within high coverage positions and exchange to another nucleotide according to Cancer type scpecific ATCG freq.
- Repreat for each nucleotide transition by # total mutation. 
- $Mutation Count_{p}$ = 
\begin{equation*} \left( \sum_{i=1}^n V_{p,i} * M_i \right)
\end{equation*}
-  Using Expected Mut Freq Matrix, compute the expected number of mutations at each position. Mutation count for each position ,$P$, is  $V_{i}$ the number of occurrence of codon $i$, at $P$, multiplied by the number of frequency, $M_i$ of Condon $i$ summed across all condons.
-  Simulation with both (Nonsyn + Syn) mutation total but scale it out later by Non-syn/Total ratio for each codon

### <font color="darkblue">Select Highly Coveraged Position of the Coding Domain Sequence(CDS) in GPCRs  </font>

```python
#Create Coding Domain Sequence(CDS) Dict
NM_dict,CDS_dict,noNMID={},{},[]
for i in list(HighCovPosi_dict.keys()):
    NM_dict[i.split('_',1)[1]]=i

faiter = (x[1] for x in groupby(open(cDNA_file), lambda line: line[0] == ">"))
for header in faiter:
    header = next(header)[1:].strip()
    seq="".join(s.strip() for s in next(faiter)).upper()
    coding_DNA = Seq(seq,IUPAC.unambiguous_dna)
    NMID=header.split('_',2)[2].split(' ')[0] 
    try:
        CDS_dict[NM_dict[NMID]]=re.findall('...',str(coding_DNA))
    except: noNMID.append(NMID) # Isoforms
print ('Total RNA in alignment :', len(CDS_dict))

def Codon_Freq(DF=None,Class=None,C=None):
    CDS_HighCov={}
    for i in DF['header'].unique():
        temp=[]
        for j in HighCovPosi_dict[i]:
            try: temp.append(CDS_dict[i][int(j)-1])
            except: temp.append('-')
            CDS_HighCov[i]=temp 
    Codon_posi_df=pd.DataFrame.from_dict(CDS_HighCov,orient='index',columns=HighCovPosi_DF.columns.tolist())
    Codon_posi_df.to_csv('../Output/%s/%s/%s_%s_Obs1_Codon_ResPosi.csv'%(C,Class,C,Class))
    
    Codon_Count=pd.DataFrame(columns=Codon_ls+['-'],index=HighCovPosi_DF.columns.tolist())
    for posi in HighCovPosi_DF.columns.tolist():
        for cd in Codon_ls+['-']:
            Codon_Count.loc[posi,cd]=Codon_posi_df[posi].tolist().count(cd)
    Codon_Count=Codon_Count.fillna(0)
    Codon_Count.to_csv('../Output/%s/%s/%s_%s_Obs2Codon_Count.csv'%(C,Class,C,Class))

    plt.figure(figsize=(16,12))
    sns.heatmap(Codon_Count,cmap='viridis')
    plt.savefig('../output/%s/%s/%s_%s_Obs3_Codon_Count_heatM.pdf'%(C,Class,C,Class),format='pdf')
    plt.close()

    Codon_Total=Codon_Count.sum(axis=0).reset_index().set_index('index').values
    Codon_Freq=Codon_Count.T/Codon_Total
    Codon_Freq.to_csv('../output/%s/%s/%s_%s_Obs4_Codon_Freq.csv'%(C,Class,C,Class))
    return (Codon_Freq)
```

### <font color="darkblue"> Compute Expected Number of Mutations </font>

```python
def Expected_Mut(DF=None,C=None,Class=None,RNAt=None):
    Missense_Count= int(DF[DF['Type']=='Missense_Mutation']['Type'].count())
    All_SNP_Count=DF['Type'].count()
    
    ATCG_MutFreq=pd.read_csv('../data/Genomics/TCGA/ATCG/%s_ATCGFreq.csv'%C,header=0,index_col=0)
    ATCG_Mut=(ATCG_MutFreq*All_SNP_Count).apply(np.round,decimals=0)
    ATCG_Mut.to_csv('../output/%s/%s/%s_%s_Exp1_ATCG.csv'%(C,Class,C,Class))
    
    Expected_Condon_Mut=[]
    for i in ['A','T','C','G']: # Ori Nucleotide
        for j in ['A','T','C','G']: # Alt Nucleotide
            if i != j:
                Count=0
                Num_Mut=int(ATCG_Mut[j][i]) # Alt, Ori
                while Count < Num_Mut:
                    Ori=random.choice(Codon_ls)
                    Codon_P=random.choice([0,1,2])
                    if Ori[Codon_P] == i :
                        Alt=list(Ori)
                        Alt[Codon_P]=j
                        Alt=''.join(Alt)
                        if Alt != 'TGA' and Alt != 'TAA' and Alt != 'TAG' : # STOP codon
                            Expected_Condon_Mut.append(Ori+'-'+Alt)
                            Count+=1
    ExpectedMut=pd.DataFrame(index=Codon_ls,columns=Codon_ls)
    for k in Codon_ls:
        for l in Codon_ls:
            ExpectedMut.loc[k,l]=Expected_Condon_Mut.count(k+'-'+l)
    ExpectedMut=ExpectedMut.fillna(0)
    ExpectedMut.to_csv('../output/%s/%s/%s_%s_Exp2_Codon.csv'%(C,Class,C,Class))
    
    # ExpectedMut HeatMap
    plt.figure(figsize=(16,12))
    sns.heatmap(ExpectedMut,cmap='viridis')
    plt.savefig('../output/%s/%s/%s_%s_Exp3_Codon.pdf'%(C,Class,C,Class),format='pdf')
    plt.close()

    Codon_MutT_dict={}
    Codon_freq_dict=ExpectedMut.to_dict('index')
    for RNA_O in Codon_freq_dict:
        Mis=0
        Silent=0
        Ori=Seq(RNA_O,generic_dna).translate()

        for RNA_A in Codon_freq_dict[RNA_O]:
            Alt=Seq(RNA_A,generic_dna).translate()
            if Ori == Alt:
                Silent = Silent + int(Codon_freq_dict[RNA_O][RNA_A])
            if Ori != Alt:
                Mis = Mis + int(Codon_freq_dict[RNA_O][RNA_A])
        Codon_MutT_dict[RNA_O]=[Silent,Mis,Silent+Mis]
    Codon_ExpMut=pd.DataFrame.from_dict(Codon_MutT_dict,orient='index',columns=['Syn','nonSyn','Total'])
    Codon_ExpMut['nonSyn_adj']=(Codon_ExpMut['nonSyn']*Missense_Count)/int(Codon_ExpMut['nonSyn'].sum())
    Codon_ExpMut['Syn_adj']=(Codon_ExpMut['Syn']*Missense_Count)/int(Codon_ExpMut['Syn'].sum())
    Codon_ExpMut.to_csv('../Output/%s/%s/%s_%s_Exp4_Codon_Sum.csv'%(C,Class,C,Class))
    
    print ('{:5}{:6}| Codon_ExpMut: Total SUM {:>6}| Syn SUM{:>6} | nonSyn SUM{:>6} |  nonSyn_adj SUM{:>6} | Observed_Mut{:>6}'.format(C,Class,Codon_ExpMut['Total'].sum(), Codon_ExpMut['Syn'].sum(),Codon_ExpMut['nonSyn'].sum(),int(Codon_ExpMut['nonSyn_adj'].sum()),Missense_Count))

    plt.figure(figsize=(9,20))
    sns.heatmap(Codon_ExpMut[['nonSyn_adj']],cmap='viridis')
    plt.savefig('../output/%s/%s/%s_%s_Exp5_Codon_Sum.pdf'%(C,Class,C,Class),format='pdf')
    plt.close()

    ExpectedMut_nonSyn=RNAt.drop(['-'])*Codon_ExpMut[['nonSyn_adj']].values
    ExpectedMut_Syn=RNAt.drop(['-'])*Codon_ExpMut[['Syn_adj']].values
    return (ATCG_Mut,Codon_ExpMut,ExpectedMut_nonSyn,ExpectedMut_Syn)
```

### <font color="darkblue"> Count Observed GPCR Mutations in TCGA by Position </font>

```python
def GPCR_Obs_Count(DF=None,C=None,Class=None):
    DF=DF.astype({'Resi':str})
    TCGA_Syn_NonSyn=DF.groupby('header')['Resi'].apply(list).to_dict()
    TCGA_Syn=DF[DF['Type']=='Silent'].groupby('header')['Resi'].apply(list).to_dict()
    TCGA_nonSyn=DF[DF['Type']=='Missense_Mutation'].groupby('header')['Resi'].apply(list).to_dict()
    
    Syn_nonSyn_dict,Syn_dict,nonSyn_dict=defaultdict(list),defaultdict(list),defaultdict(list)
    for G in TCGA_Syn_NonSyn:
        Syn_nonSyn_dict[G] = dict(Counter(list(map(Posi_queryMatch[G].get,TCGA_Syn_NonSyn[G]))))
    for G in TCGA_Syn:
        Syn_dict[G]        = dict(Counter(list(map(Posi_queryMatch[G].get,TCGA_Syn[G]))))
    for G in TCGA_nonSyn:
        nonSyn_dict[G]     = dict(Counter(list(map(Posi_queryMatch[G].get,TCGA_nonSyn[G]))))
    
    Syn_NonSyn_MUT=pd.DataFrame.from_dict(Syn_nonSyn_dict,orient='index').fillna(0)
    Syn_MUT       =pd.DataFrame.from_dict(Syn_dict,orient='index').fillna(0)
    nonSyn_MUT    =pd.DataFrame.from_dict(nonSyn_dict,orient='index').fillna(0)
   
    # By Position
    Syn_NonSyn_MUT_posi =Syn_NonSyn_MUT.T.sum(axis=1).reset_index().set_index('index').rename(columns={0:'Syn_nonSyn'})
    Syn_MUT_posi        =Syn_MUT.T.sum(axis=1).reset_index().set_index('index').rename(columns={0:'Syn'})
    nonSyn_MUT_posi     =nonSyn_MUT.T.sum(axis=1).reset_index().set_index('index').rename(columns={0:'nonSyn'})

    Obs_MUT_posi=pd.concat([Syn_NonSyn_MUT_posi,Syn_MUT_posi,nonSyn_MUT_posi],axis=1,ignore_index=False,sort=True).fillna(0).reset_index().rename(columns={'index':'Residue'}).astype({'Residue':'int32'}).sort_values(by='Residue')
    Obs_MUT_posi['Generic']=Obs_MUT_posi['Residue'].map(Numbering)
    Obs_MUT_posi.to_csv('../output/%s/%s/%s_%s_Obs5_Position.csv'%(C,Class,C,Class),index=False)
    
    ### Histogram
    sns.set_style("white")
    figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='w')
    _, bins, _ = plt.hist(Obs_MUT_posi['Syn'].tolist(), bins=12,alpha=0.9, density=False,color='k',ec='white',linewidth=1.5,label='Syn')
    _ = plt.hist(Obs_MUT_posi['nonSyn'].tolist(), bins=bins, alpha=0.7, color='red', ec='white',linewidth=1.5,density=False,label='Non-Syn')

    plt.title('%s Cancer | Synonymous and Non-synonymous Mutation Distribution'%(C))
    plt.xlabel('Mutations per Position')
    plt.ylabel('Mut Freq')
    sns.despine(left=False, bottom=False, right=True)
    plt.legend()
    plt.savefig('../output/%s/%s/%s_%s_Obs5_Position_Hist.pdf'%(C,Class,C,Class),format='pdf')
    plt.close()
    
    # By Protein
    Syn_NonSyn_MUT_protein =DF['Gene'].value_counts().reset_index().rename(columns={'index':'Gene','Gene':'Syn_nonSyn'}).set_index('Gene',drop=True)
    Syn_MUT_protein        =DF[DF['Type']=='Silent']['Gene'].value_counts().reset_index().rename(columns={'index':'Gene','Gene':'Syn'}).set_index('Gene',drop=True)
    nonSyn_MUT_protein     =DF[DF['Type']=='Missense_Mutation']['Gene'].value_counts().reset_index().rename(columns={'index':'Gene','Gene':'nonSyn'}).set_index('Gene',drop=True)
    Obs_MUT_protein=pd.concat([Syn_NonSyn_MUT_protein,Syn_MUT_protein,nonSyn_MUT_protein],axis=1,ignore_index=False ,sort=True) 
    Obs_MUT_protein=Obs_MUT_protein.fillna(0)
    Obs_MUT_protein.to_csv('../output/%s/%s/%s_%s_Obs5_Protein.csv'%(C,Class,C,Class))
    return (Obs_MUT_protein,Obs_MUT_posi)
```
