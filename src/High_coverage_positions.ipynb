```python
import pandas as pd
from IPython.display import display, HTML
from itertools import groupby
```

```python
#Class A GPCR ID Table
GPCR_ID=pd.read_csv('../data/GPCR/GPCRlist.csv',usecols=['Class','Family.name','HGNC.symbol','Human.nucleotide.RefSeq'])
GPCR_classA_ID=GPCR_ID[GPCR_ID['Class']=='A'].drop_duplicates(subset=['HGNC.symbol'],keep='first')
print ('The number of class A GPCR : %s'%GPCR_classA_ID.shape[0])
display(HTML(GPCR_classA_ID.head(5).to_html()))

ClassA_GPCR=GPCR_classA_ID['HGNC.symbol'].tolist()
NM_dict=GPCR_classA_ID.set_index('HGNC.symbol')['Human.nucleotide.RefSeq'].to_dict()

faiter = (x[1] for x in groupby(open('../data/GPCR/GPCRseq.fasta'), lambda line: line[0] == ">"))
Outf=open('../data/GPCR/ClassAGPCRseq.txt','w')
IDtemp=[]
for header in faiter:
    header = next(header)[1:].strip()             # drop the ">"
    seq ="".join(s.strip() for s in next(faiter)) # join all sequence lines to one.
    ID=header.split('GN=')[1].split(' ')[0]
    if ID in ClassA_GPCR:
        IDtemp.append(ID)
        Outf.write('>'+ID+'_'+NM_dict[ID]+'\n'+seq+'\n')
```

## <font color="darkblue"> Select Class A Position</font>
- Select Residue positions with MSA coverage of 90% from All class A GPCR
- Query protein : ADRB2_NM_000024
- Header e.g. : Gene_NMID

```python
Align_Pos,POI_ls, highcovarray, High_Cov_dict={},[],[],{}
faiter = (x[1] for x in groupby(open('../data/GPCR/ClassAGPCRseq_PROMALS_aln.fa'), lambda line: line[0] == ">"))
for header in faiter:
    header = next(header)[1:].strip()
    seq = list("".join(s.strip() for s in next(faiter)))
    Lentotalseq=len(seq)
    Align_Pos[header]=seq
    POI_ls.append(header)
# Count Residue Position
for p in Align_Pos:
    pos_count=0
    for i in range(0,Lentotalseq,1):
        if Align_Pos[p][i] != '-':
            pos_count=pos_count+1
            Align_Pos[p][i]=pos_count
# Select Highly Coveraged GPCR Position 
totalcounts=[0]*Lentotalseq
for i in Align_Pos:
    innerloop=0
    for j in Align_Pos[i]:
        if j != '-':
                tempcount=totalcounts[innerloop]+1
                totalcounts[innerloop]=tempcount
        innerloop=innerloop+1
C=0
for k in totalcounts:
    if k >= 0.9 * len(Align_Pos):
        try:
            highcovarray.append(C) 
        except:
            highcovarray=[C,]        
    C=1+C

for G in POI_ls:
    temp=[]
    for m in highcovarray:
        temp.append(Align_Pos[G][int(m)])
        High_Cov_dict[G]=temp
Position_df=pd.DataFrame.from_dict(High_Cov_dict,orient='index',columns=High_Cov_dict['ADRB2_NM_000024']).reset_index().drop(['-'],axis=1)
display(HTML(Position_df.head(10).to_html()))
Position_df.to_csv('../data/GPCR/ClassAGPCR_HigCovPosi.csv',index=False)
print ("Number of Position Selected: ",len(Position_df.columns.tolist())-1) # exclude index column
print ("Number of proteins: ",Position_df['index'].count())
```
